## 원핫인코딩
```
글자였을때 못하는 사칙연산이 가능하므로 위와같은 작업은 중요함
원핫인코딩이 위 작업과 차이점= 0or1로만 표현
원핫인코딩 활용예)
해당변수가 미국(1)인경우 두개 수가 합쳐져서 영국(2)가 되는 오류가 없도록 
dummy번수를 만들어 관리

```

````python
 #새로운파일불러옴
    df_n = pd.read_csv('brand_ad_ex2.csv')
````

````python
df_n.head()
````

````python
new_col=['adv','brand','start','end','imp','clk','cmp','ind']
df_n.columns = new_col
df_n.loc[0:1]
````

````python
df_n['ind'].value_counts()
````

````python
# . value_counts() 와 . count()의 차이점
df_n['ind'].count()
# . value_counts() = 해당컬럼의 데이터항목과 각 항목의 갯수, 데이터타입까지 다보여줌
# . count() =  해당칼럼의 데이터 총갯수만 보여줌
````

````python
# df_n['ind'].value_counts()의 내용을 원핫인코딩 데이터테이블로 뽑아냄
dummies = pd.get_dummies(df_n['ind'])
dummies
````

````python
#생성된 더미데이터에서 건설업(첫째컬럼)만 지우려면 
dummies = pd.get_dummies(df_n['ind'], drop_first=True)
dummies.head()
````

기존데이터에 더미변수 추가하는방법
## join

````python
df2= df_n.copy()
df2 = df2.join(dummies)
df2
````
