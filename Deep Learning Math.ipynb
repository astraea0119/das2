# 1차함수 linaer_funcion

```python
#define=함수명령 넘겨받는 파라메타이기 떄문에 소문자x로 표기
def linaer_funcion(x):
    a = 0.5 #기울기
    b = 2 #절편
    
    return a * x + b #단순한 1차함수식으로 표현가능
```

```python
print(linaer_funcion(5))
#리니어펑션호출해서 지정한 x값(스칼라)을 1차함수식에 대입시킴
```

# 2차함수 quandratic_function

```python
def quandratic_function(x):
    a = 1
    b = -2
    c = -2
    return a*x**2+b*x+c #x**2=x의 2제곱
```

```python
y= quandratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('quandratic_function')

```

# 3차함수 cubic_function

```python
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    
    return a*x**3+b*x**2+c*x+d
```

```python
y= cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('cubic_function')
```

# 신경망구조
신경망에서는 함수의 절편을 [편향치]로 부름

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
#시각화(그래프) 설정
```

```python
#앤드gate 생성(x1, x2)=입력값
def AND(x1, x2):
    input = np.array([x1, x2]) #입력값
    weights= np.array([0.4,0.4]) #가중치설정
    bias = -0.6 #편향치
    
    value = np.sum(input * weights) + bias
    #조건설정
    if value <=0:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
#그래프화
x1 = np.arange(-2, 2, 0.01) #-2~2까지 표시. 표시단위0.01
x2 = np.arange(-2, 2, 0.01)
bias = -0.6

y = (-0.4* x1 - bias) /0.4 
# (-0.4=가중치* x1) /0.4 

#축설정
plt.axvline(x=0) #v=버티컬 수직
plt.axhline(y=0) #h=호라이즌 수평

plt.plot(x1,y,'r--')
```

###### 그래프해석= 
###### 빨간점선=값을 분류하는 기준이 되는 선
###### 주황점, 검은삼각=andgate표현

```python
# OR게이트 생성  
#or가 되기 위한 조건=and게이트 빨간선보다 편향치를 낮춰서 0,0제외한 전체수가 빨간선위로 포함되게 해야함
# 위의 조건을 적용하기 위해서 편향치 하향조정
def OR(x1, x2):
    input = np.array([x1, x2]) #입력값
    weights= np.array([0.4,0.4]) #가중치설정
    bias = -0.3 #편향치
    
    value = np.sum(input * weights) + bias
    #조건설정
    if value <=0:
        return 0
    else:
        return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
#동일한 세포에서 바이어스(편향치)조정으로 and->or로 바뀔 수 있음
#bias의 역할=기준선 상하 조정
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = -0.3
y = (-0.4 * x1 - bias) / 0.4
plt.axvline(x=0)
plt.axhline(y=0)
plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
#마커수정가능
```

```python
#NAND (not and)게이트생성
def NAND(x1, x2):
    input = np.array([x1, x2]) #입력값
    weights= np.array([0.4,0.4]) #가중치설정
    bias = -0.7 #편향치
    
    value = np.sum(input * weights) + bias
    #조건설정
    if value <=0:
        return 0
    else:
        return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
#결과 오류
#NAND게이트는 가중치, 편향치 둘다 조정해야함
def NAND(x1, x2):
    input = np.array([x1, x2]) #입력값
    weights= np.array([-0.6,-0.6]) #가중치설정
    bias = 0.7 #편향치
    
    value = np.sum(input * weights) + bias
    #조건설정
    if value <=0:
        return 0
    else:
        return 1
    
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

위의 AND, OR, NAND 사용결과 0110 배열을 출력 못함<p>
AI의 한계도래 -> 다중 디셉트콘 **XOR**발표 by.제프리힌턴

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)

  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

```python
from keras import models
from keras import layers
#

#신경망생성 순차적 진행명령 코드sequential()
network = models.Sequential()
#네트워크라는 변수로 저장
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,))) #인풋오타
# add() 네트워크라는 변수에 신경망 추가
##신경망추가명령 layers.Dense=덴스레이어추가(512=뉴런수,activation활성함수='relu', 
##이 활성함수로 들어오는 입력값=imput_shape= 입력모양=(28*28, 콤마= 한차원이 더있음을 알림)))

network.add(layers.Dense(10,activation='softmax'))
```

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy']) #네트워크, 컴파일, 쉼표 오탈자

#네트워크 컴파일='신경망을 완성시킨다' 선언문
```

```python
#신경망에 맞게 이미지를 바꿔줌
train_images = train_images.reshape(60000, 28 * 28 )
#reshape=넘파이에서 자료의 구조를 바꿀때 사용 6만,28,28-> 6만,28*28 로 구조변경 셀을 784개 짜리의1차원이미지로 만듬 이 데이터가 6만개
train_images.shape
```
